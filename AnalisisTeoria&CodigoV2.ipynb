# Método de Newton-Raphson para sistemas de ecuaciones no lineales

El método se basa en usar una recta tangente a la gráfica de f para aproximar esta gráfica, cerca del punto donde la función se anula. 

Por ejemplo supongamos que se tiene la aproximación xk a la raíz de F(x). Se traza la recta tangente a la curva en el punto (xk, F(xk)) y se hace uso de la fórmula: 

\begin{equation*}
 y= F(xk) + F'(xk)(x-xk)  
\end{equation*}

La recta encontrada cruza el eje x en un punto xk+1, la cual será la nueva aproximación para la raíz.

\begin{equation*}
Imagen 1.Grafica: funcion,rectas
\end{equation*}

El punto donde la recta corta el eje x es dado por la ecuación:

\begin{equation*}
 0= F(xk) + F'(xk)(xk+1-xk)   
\end{equation*}

Por lo que el método de Newton Raphson define las aproximaciones a la raíz como:

\begin{equation*}
xk+1 =xk-(f(xk)/f'(xk)), donde: k= 0,1,2,3,4…..
\end{equation*}

todo a partir de una aproximación inicial de x0, la cual siempre será dada en el ejercicio y siempre que f'(xk) ≠ 0

## El método de Newton 
Este método consiste en una generalización del método de Newton Raphson, el cual permite trabajar de manera fácil sistemas de ecuaciones NO lineales. 

Para resolver el sistema de ecuaciones se tiene que f(x)=0; donde x=(x1,...,xn)t

Haciendo uso del desarrollo de Taylor para varias variables f en base a una aproximación xk=(xk1,...,xkn)t, tenemos:

\begin{equation*}
0 = f(xk)+Df(xk)(-xk)
\end{equation*}
Donde  es la solución del sistema de ecuaciones y Df(xk) es la matriz Jacobiana 

\begin{equation*}
Imagen 2. Matriz Jacobiana
\end{equation*}

Haciendo uso de la transpuesta de la matriz tenemos:
\begin{equation*}
= (xk)-Df(xk) -1* f(xk)
\end{equation*}

Como el método lo indica se inicia dada la aproximación (xk), luego se toma la aproximación (xk+1) y se sustituye en la ecuación:
\begin{equation*}
xk+1= (xk)-Df(xk) -1* f(xk)
\end{equation*}

## Ventajas del Método de Newton

Una de las ventajas del método de Newton–Raphson además de su velocidad de convergencia, es que
se puede generalizar fácilmente a sistemas de ecuaciones no lineales.

## Desventajas del Método de Newton

-Lenta convergencia debida a la naturaleza de una función en particular.

-No existe un criterio general de convergencia.

-Cuando un punto de inflexión, f’’(x) = 0, ocurre en la vecindad de una raíz.

##  CODIGO 1 EJEMPLO ECUACION NO LINEAL
### Encontrar las raices de:

\begin{equation*}
x^2   = 4
\end{equation*}

Para el metodo de Newton tendremos que saber la funcion y su derivada, para este caso en particular tendremos:

\begin{equation*}
F(x)= x^2 - 4
\end{equation*}

\begin{equation*}
F'(x)= 2x
\end{equation*}

Con eso lograremos obtener esta formula

\begin{equation*}
X ^ {(k+1)} = x ^{(k)}- \frac{(x^{(k)})^{(2)} - 4}{2x ^{(k)}}
\end{equation*}


```python
import numpy as np

#Definimos la funcion a trabajar
def f(x):
    return x**2 - 4

#Derivada de f
def df(x):
    return 2*x

#Definimos el numero maximo de iteraciones 
i_max= 20

#Definimos una estimacion inicial
x= 1.5

#Definimos el residual

res = abs(f(x))

i=0
while(res > 1e-8):
    print( i, x, f(x) )
    
    #Se ejecuta la formula para el metodo de Newton
    x = x - f(x)/df(x)
    
    #Se actualiza el residuo
    res= abs(f(x))
    
    #Actualizamos el contador 'i'
    i+=1
    
    if(i > i_max):
        print('La interacion no converge')
        break
```

# Problema Codigo 1 actualizado para un sistema de ecuacion no lineal 
## Tenemos el siguiente sistema de ecuaciones

\begin{equation*}
x^2 + y^ 2 = 1
\end{equation*}

\begin{equation*}
\frac{x^2}{(\frac{3}{2})^2} + \frac{y^2}{(\frac{1}{2})^2} = 1
\end{equation*}

### Con este sistema de ecuaciones tendremos la siguiente grafica:



\begin{equation*}
Imagen 3.Grafica del sistema de ecuacion
\end{equation*}

### Al ser la interseccion donde estos valores se igualan, para este ejercicio se tendria 4 soluciones

#### Pondremos las ecuaciones en forma de funcion para eso se igualan a 0


\begin{equation*}
f1(x,y) = x^2 = y^2 -1 = 0
\end{equation*}

\begin{equation*}
f2(x,y) = \frac{4}{9}x^2 + 4y^2 -1 = 0 
\end{equation*}



#### Pondremos el sistema de ecuaciones en forma de campo vectorial

##### Definimos nuestras incognitas

\begin{equation*}
x =  \begin{vmatrix}
\mathbf{x}  \\
\mathbf{y}  
\end{vmatrix}
\end{equation*}

##### Definimos nuestra funcion F(X) con dos componentes f1 y f2; para la cual buscamos valores de 'x' y 'y' que nos devuelva '0'

\begin{equation*}
F(x) =  \begin{vmatrix}
\mathbf{x^2 + y^2 -1 }  \\
\mathbf{\frac{4}{9}x^2 + 4y^2 -1}  
\end{vmatrix} = \begin{vmatrix}
\mathbf{0}  \\
\mathbf{0}  
\end{vmatrix}
\end{equation*}

##### Definimos la funcion Jacobiana

\begin{equation*}
\frac{\partial F}{\partial X} =  \begin{vmatrix}
\frac{\partial f1}{\partial x} &  \frac{\partial f1}{\partial y}  \\
\frac{\partial f2}{\partial x} &  \frac{\partial f2}{\partial y} 
\end{vmatrix} = \begin{vmatrix}
\mathbf{2x} & \mathbf{2y}  \\
\mathbf{\frac{8}{9}x} & \mathbf{8y}  
\end{vmatrix}
\end{equation*}


```python
import numpy as np

#Definimos la funcion a trabajar
def f(x):
    f1=x[0]**2+x[1]**2-1
    f2=4*x[0]**2/9+4*x[1]**2-1
    return np.array([f1,f2])


#Derivada de f
def df(x):
    return np.array([[2*x[0],2*x[1]], [8*x[0]/9,8*x[1]]])

#Definimos el numero maximo de iteraciones 
i_max= 100

#Definimos una estimacion inicial; De esto dependera la solucion encontrada, segun la estimacion inicial se podria encontrar cualquiera de las 4 soluciones
x= np.array([1,1])

                     
for k in range(i_max):
    
    #Almacenamos los valores de 'x'
    xold=x
    
    #Definimos la Jacobiana inversa
    Jinv=np.linalg.inv(df(x))
    
    #Formula de Newton Raphson para sistema de ecuaciones no lineales 
    x=x-np.dot(Jinv,f(x))
    
    #Definimos para calcular error
    e=np.linalg.norm(x-xold)
    
    print(k,x,e)
    
    if(e <1e-10):
        break
```
